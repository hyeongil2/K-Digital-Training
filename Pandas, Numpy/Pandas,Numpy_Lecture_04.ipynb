```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]


plt.plot(x, y)
plt.title("First Plot")
plt.xlabel("x")
plt.ylabel("y")
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("First Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
fig.set_dip(300)
fig.savefig("first_plot.png")
```

```python
fig, ax = plt.subplots()
x = np.arange(15)
y = x ** 2
ax.plot(
x, y,
linestyle = ":",
marker = "*",
color = "#524FA1"
)
```

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, linestyle = "-")
# solid
ax.plot(x, x*2, linestyle = '--')
# dashed
ax.plot(x, x*4, linestyle = '-.')
# dashdot
ax.plot(x, x*6, linestyle = ":")
# dotted
```

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, color = "r")

ax.plot(x, x*2, color = "green")

ax.plot(x, x*4, color = "0.8")

ax.plot(x, x*6, color = "#524FA1")
```

```python
x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots()

ax.plot(x, np.sin(x))

ax.set_xlim(-2, 12)

ax.set_ylim(-1.5, 1.5)
```

```python
# 
fig, ax = plt.subplots()

ax.plot(x, x, label = 'y=x')

ax.plot(x, x**2, label  ='y=x^2')

ax.set_xlabel("x")

ax.set_ylabel("y")

ax.legend(
loc = 'upper right', # 
shadow = True, # 
fancybox = True, 
borderpad = 2 # 
)
```

```python
# Scatter
fig, ax = plt.subplots()

x = np.arange(10)

ax.plot(
x, x**2, "o",
markersize = 15,
markerfacecolor = 'white',
markeredgecolor = "blue")
```

```python
# Scatter

fig, ax = plt.subplots()

x = np.random.randn(50)

y = np.random.randn(50)

colors = np.random.randint(0, 100, 50)

sizes = 500 * np.pi * np.random.rand(50) ** 2
ax.scatter(
x, y, c = colors, s=sizes,
alpha = 0.3 # 
)
```

```python
# Bar plot
x = np.arange(10)
fig, ax = plt.subplots(figsize = (12, 4)) #  
ax.bar(x, x*2)
```

```python
# Bar plot
x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)

data = [x, y , z]

fig, ax = plt.subplots()
x_ax = np.arange(3)
for i in x_ax:
    ax.bar(x_ax, data[i],
          bottom=np.sum(data[:i], axis =0))
ax.set_xticks(x_ax)
ax.set_xticklabels(["A", "B", "C"])
```

```python
#Histogram
fig,ax = plt.subplots()
data = np.random.randn(1000) # rand = random nomial
ax.hist(data, bins=50)
```
